```python
import numpy as np
import matplotlib as mpl

mpl.use('agg')

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
all_means_summary = []
all_means_hist = []
```

```python
lines = []
#files = ['f4-4-10000_0.01_MAPE_summary.txt', 'f4-4-10000_0.05_MAPE_summary.txt', 'f4-4-10000_0.1_MAPE_summary.txt',
#        'f4-4-10000_0.5_MAPE_summary.txt', 'f4-4-10000_MAPE.txt']
files = ['results/f10-4-1000_2_10_MSE_summary.txt', 
         'results/f10-4-1000_2_10_MSE.txt']
for f_str in files:
    with open(f_str,'r') as f:
        lines.append(f.readlines())
```

```python
all_values = []
for line in lines:
    values = []
    for l in line:
        values.append(np.array(map(float,l.split(','))).mean())
    values = np.array(values)
    all_values.append(values)
```

```python
#all_values[0] = all_values[0][:10]
#all_values[1] = all_values[1][:10]
```

```python
all_values[1] = all_values[1][~np.isnan(all_values[1])]
```

```python
all_values = []
for line in lines:
    values = []
    for l in line:
        values.append(np.array(map(float,l.split(','))).mean())
    values = np.array(values)
    all_values.append(values)
```

```python
all_means_summary.append(all_values[0].mean())
all_means_hist.append(all_values[1].mean())
```

```python
print(all_means_hist)
```

```python
fig = plt.figure(1, figsize=(9, 6))

ax = fig.add_subplot(111)

bp = ax.boxplot(all_values)
ax.set_xticklabels(['summary', 'histogram'])

ax.set_ylabel('MSE')
ax.set_title('MSE for federated with different distributions - 1000 samples')

plt.show()
#fig.savefig('fig1.png', bbox_inches='tight')
```

```python
lines = []
#files = ['f4-4-10000_0.01_MAPE_summary.txt', 'f4-4-10000_0.05_MAPE_summary.txt', 'f4-4-10000_0.1_MAPE_summary.txt',
#        'f4-4-10000_0.5_MAPE_summary.txt', 'f4-4-10000_MAPE.txt']
files = ['results/0_3_60_normal_normal_normal/f10-6-1000_3_10_time_summary.txt', 
         'results/0_3_60_normal_normal_normal/f10-6-1000_3_10_time.txt']
for f_str in files:
    with open(f_str,'r') as f:
        lines.append(f.readlines())
```
